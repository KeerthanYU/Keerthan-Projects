```python
numerical_cols = df.select_dtypes(include=['int64','float64']).columns
df[numerical_cols].describe()
```

```python
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"{col} value counts:\n{df[col].value_counts()}\n")
```

```python
df.isnull().sum()
```

```python
for col in numerical_cols:
    plt.figure()
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
    plt.savefig(f'C:/Users/KEERTHAN/OneDrive/Documents/GitHub/Keerthan-Projects/Python Project 3/Student_analyser/reports/figures{col}_hist.png')
    plt.close()
```

```python
for col in categorical_cols:
    plt.figure()
    sns.countplot(x=col, data=df)
    plt.title(f'Counts of {col}')
    plt.xticks(rotation=45)
    plt.savefig(f'C:/Users/KEERTHAN/OneDrive/Documents/GitHub/Keerthan-Projects/Python Project 3/Student_analyser/reports/figures{col}_count.png')
    plt.close()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.savefig('C:/Users/KEERTHAN/OneDrive/Documents/GitHub/Keerthan-Projects/Python Project 3/Student_analyser/reports/figures/correlation_matrix.png')
plt.close()
```

```python
for cat_col in categorical_cols:
    for num_col in numerical_cols:
        plt.figure()
        sns.boxplot(x=cat_col, y=num_col, data=df)
        plt.title(f'{num_col} vs {cat_col}')
        plt.xticks(rotation=45)
        plt.savefig(f'C:/Users/KEERTHAN/OneDrive/Documents/GitHub/Keerthan-Projects/Python Project 3/Student_analyser/reports/figures/{num_col}_vs_{cat_col}_box.png')
        plt.close()
```

```python
for col in numerical_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.savefig(f'C:/Users/KEERTHAN/OneDrive/Documents/GitHub/Keerthan-Projects/Python Project 3/Student_analyser/reports/figures/{col}_boxplot.png')
    plt.close()
```
