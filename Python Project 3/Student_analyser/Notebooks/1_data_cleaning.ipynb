```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler,LabelEncoder
```

```python
df=pd.read_csv('C:/Users/KEERTHAN/OneDrive/Documents/GitHub/Keerthan-Projects/Python Project 3/Student_analyser/data/student-por.csv')
df.head()
```

```python
df.info()
df.describe(include="all")
```

```python
df.columns=(
    df.columns.str.strip()
            .str.lower()
            .str.replace(" ","_")
            .str.replace("-","_")
)
df.columns
```

```python
df.isnull().sum()
df.duplicated().sum()
df=df.drop_duplicates()
```

```python
df['school']=df['school'].str.lower().str.strip()
```

```python
df['sex']=df['sex'].str.lower()
```

```python
df['age'].dtype
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df = df[df['age'].between(15, 22)]
```

```python
df['address']=df['address'].str.upper().str.strip()
```

```python
df['pstatus'] = df['pstatus'].str.lower()
```

```python
df['medu'] = pd.to_numeric(df['medu'], errors='coerce')
df['fedu'] = pd.to_numeric(df['fedu'], errors='coerce')

df = df[df['medu'].between(0,4)]
df = df[df['fedu'].between(0,4)]
```

```python
for col in ['mjob','fjob']:
    df[col] = df[col].str.lower().str.strip()
```

```python
df['reason'] = df['reason'].str.lower().str.strip()
```

```python
df['guardian'] = df['guardian'].str.lower().str.strip()
```

```python
df = df[df['traveltime'].between(1,4)]
```

```python
df = df[df['studytime'].between(1,4)]
```

```python
df = df[df['failures'].between(0,3)]
```

```python
binary_cols = ['schoolsup','famsup','paid','activities','nursery',
               'higher','internet','romantic']

for col in binary_cols:
    df[col] = df[col].str.lower().str.strip().map({'yes':1,'no':0})
```

```python
scale_cols = ['famrel','freetime','goout','dalc','walc','health']

for col in scale_cols:
    df = df[df[col].between(1,5)]
```

```python
df = df[df['absences'].between(0, 93)]
```

```python
for col in ['g1','g2','g3']:
    df = df[df[col].between(0,20)]
```

```python
df = df[(df['g1'] <= df['g2']) & (df['g2'] <= df['g3'])]
```

```python
num_cols = ['age','traveltime','studytime','failures','famrel',
            'freetime','goout','dalc','walc','health',
            'absences','g1','g2','g3']

def remove_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)]

for col in num_cols:
    df = remove_outliers_iqr(df, col)
```

```python
le = LabelEncoder()

cat_cols = ['school','sex','address','famsize','pstatus',
            'mjob','fjob','reason','guardian']

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df['alcohol_total'] = df['dalc'] + df['walc']
```

```python
df['grade_variation'] = df[['g1','g2','g3']].std(axis=1)
```

```python
df['performance'] = pd.cut(df['g3'],
                           bins=[0,10,15,20],
                           labels=['poor','average','excellent'])
```

```python
scaler = StandardScaler()

df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df['sex'] = df['sex'].map({0:'female', 1:'male'})
```
